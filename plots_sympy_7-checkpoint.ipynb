```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
```

```python
x = np.arange(-5, 5, 0.01)
y1 = x + 1
y2 = -x + 1
fig, (ax) = plt.subplots(1, 1, sharex=True, figsize=(5, 3))

plt.xlim(-10,10)
plt.ylim(-20,20)
ax.fill_between(x , 3*y1 , 3*y2)
ax.set_title('fill between y1 and y2')
ax.set_xlabel('x')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4*np.pi, 0.01) 

y1 = np.sin(x)*np.sin(x)
y2 = np.cos(x)*np.cos(x)

plt.xlim(-0,10)
plt.ylim(-5,5)

plt.plot(x, y1, label='$y1 = sin^2(x)$')
plt.plot(x, y2, label='$y2 = cos^2(x)$')

plt.fill_between(x , y1 - 2 , y2 - 2,color = 'yellowgreen',label='Area_fill_y1,y2')

plt.legend()

plt.grid()
plt.show()
```
