```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi , sin ,cos
```

```python
q = 30*pi/180
dq = 0.01
x=[]
y=[]
x_1=np.linspace(-10,10,650)
y_1=[]
for i in range(650):
    q = q + dq
    x.append(3*cos(q))
    y.append(10*sin(q))
    y_1.append(200-x_1[i]**2)

fig, (ax) = plt.subplots(figsize=(7, 8))

ax.fill(x, y, "b", x_1, y_1, "r",alpha=0.5)  # 2 figuras
ax.grid()
plt.show()
```

```python
q = 30*pi/180
dq = 0.01
x=[]
y=[]
x_1=np.linspace(-10,10,650)
y_1=[]
for i in range(650):
    q = q + dq
    x.append(3*cos(q))
    y.append(10*sin(q))
    y_1.append(200-x_1[i]**2)

fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, 1, figsize=(7, 8))

ax1.fill(x, y, facecolor='lightblue', edgecolor='orangered', linewidth=3)
ax1.grid()

ax2.fill_between(x_1, x, facecolor='yellowgreen', edgecolor='orangered', linewidth=3)
ax2.grid()

ax3.fill_between(x_1,y_1, facecolor='magenta', edgecolor='k', linewidth=3,alpha=0.5)
ax3.grid()

ax4.fill_between(x_1,x_1, facecolor='green', edgecolor='k', linewidth=3,alpha=0.5)
ax4.grid()

ax5.fill_between(x, 2, facecolor='cyan', edgecolor='k', linewidth=3,alpha=0.5)
ax5.grid()

plt.tight_layout() 
plt.show()
```
