```python
#IMPORTAMOS LAS LIBRERÍAS QUE NECESITAMOS

import numpy as np
import matplotlib.pyplot as plt
from numpy import pi,cos,sin

# VALORES COORDENADAS 

tiempo = np.arange(-10, 10, 0.01)
current1 = 5*cos((pi/3)*tiempo)
current2 = 5*sin((pi/3)*tiempo)

#ESTABLECEMOS LOS LIMITES DE LOS EJES COORDENADOS

plt.xlim(-10,10)
plt.ylim(-10,10)

#DIBUJAMOS LOS GRÁFICOS SEPARADAMENTE
plt.subplot(2,1,1)

plt.plot(tiempo, current1, color='green')
plt.ylabel('current [A]')
plt.xlabel('tiempo [s]')
plt.title('AC CURRENT1')
plt.grid()

plt.subplot(2,1,2)

plt.plot(tiempo, current2, color='orange')
plt.ylabel('current [A]')
plt.xlabel('tiempo [s]')
plt.title('AC CURRENT2')
plt.grid()

# COMANDO PARA GURDAR ESPACIO EN BLANCO ENTRE LOS GRÁFICOS
plt.tight_layout() 

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, figsize=(7, 5))

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)

ax[0].plot(x, y)
ax[0].set_title("curva_1")
ax[0].grid()
ax[1].plot(x, y)
ax[1].plot(x * 3.0, y)
ax[1].set_title("curvas_2")
ax[1].grid()
plt.tight_layout() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, figsize=(7, 5))

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sinc(x)

ax[0].plot(x, y,'g.')
ax[0].plot(x, y,'k-')
ax[0].set_title("Single curve")
ax[0].grid()
ax[1].plot(x, y,'r--')
ax[1].plot(x * 3.0, y,'b-')
ax[1].set_title("Two curves")
ax[1].grid()
plt.tight_layout() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from math import cos,sin,pi

#reservamos los ejes
fig, ax = plt.subplots()

#definimos la figura
#tipo de curva

curva, = ax.plot([], [], '-',color = 'blue')

#los arreglos para sus coordenadas 
xdata, ydata = [], []

def init():
    ax.set_xlim(-10,10)
    ax.set_ylim(-15,15)
    return curva,

T = 15.5 # periodo 

def update(i):
    while i <= T:
        ax.set_title(str(round(i,1)))
        xdata.append(5*cos(2*pi*i/T))
        ydata.append(5*sin(2*pi*i/T))
        curva.set_data(xdata, ydata)
        return curva,

ani = FuncAnimation(fig, update,np.linspace(0,30,1000),
                    init_func=init, interval = 10, blit=True,repeat = False)

plt.axis("equal")

plt.grid()

plt.show()

#anim.save('Dibujos_01.gif',writer='pillow',fps=10)
```
